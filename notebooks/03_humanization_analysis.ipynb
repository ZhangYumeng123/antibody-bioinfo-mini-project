# 抗体人源化程度 PCA 分析

本 Notebook 演示：
- 用 scikit-learn 做 PCA 降维
- 可视化不同来源抗体的特征聚类
- 分析哪些特征对人源化最敏感

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import f_classif
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Day 7
from sklearn.datasets import load_wine

wine = load_wine()
df = pd.DataFrame(wine.data, columns=wine.feature_names)
df['class'] = wine.target_names[wine.target]

# PCA
X = df[wine.feature_names]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

df_plot = pd.DataFrame({
    "PC1": X_pca[:, 0],
    "PC2": X_pca[:, 1],
    "Class": df["class"]
})

# 画图
plt.figure(figsize=(8, 6))
for cls in df_plot["Class"].unique():
    subset = df_plot[df_plot["Class"] == cls]
    plt.scatter(subset["PC1"], subset["PC2"], label=cls, alpha=0.7)
plt.xlabel(f"PC1 ({pca.explained_variance_ratio_[0]:.1%})")
plt.ylabel(f"PC2 ({pca.explained_variance_ratio_[1]:.1%})")
plt.legend()
plt.title("PCA降维可视化")
plt.show()

# 敏感特征
f_scores, _ = f_classif(X_scaled, wine.target)
for name, score in sorted(zip(wine.feature_names, f_scores), key=lambda x: -x[1]):
    print(f"{name}: {score:.1f}")
```

## 小结
PCA 将高维特征压缩到2维，不同类别在特征空间中的分布一目了然。F值分析揭示了哪些特征对分类贡献最大。
